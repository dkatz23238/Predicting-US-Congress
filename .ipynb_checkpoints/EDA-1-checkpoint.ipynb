```python
import pandas as pd
import numpy as np
import seaborn as sns

def clean_party(party):
    if party == "DEM" or party == "REP":
        return party
    else:
        return "OTHER/INDIE"

df = pd.read_csv("train_set.csv")
df.CAND_PTY_AFFILIATION = df.CAND_PTY_AFFILIATION.apply(clean_party)




del df["Unnamed: 0"]
del df["SPEC_ELECTION"]
del df["PRIM_ELECTION"]
del df["RUN_ELECTION"]

log_cols = [
    "TTL_RECEIPTS",
    "TTL_DISB",
    "COH_BOP",
    "COH_COP",
    "CAND_CONTRIB",
    "CAND_LOANS",
    "OTHER_LOANS",
    "CAND_LOAN_REPAY",
    "DEBTS_OWED_BY",
    "TTL_INDIV_CONTRIB",
    "INDIV_REFUNDS",
    "CMTE_REFUNDS",

]

for col in log_cols:
    df[col] = df[col].apply(np.log1p)

pairplot_cols = [
    "CAND_PTY_AFFILIATION",
    "GEN_ELECTION_PRECENT"
    
]
```

```python
sns.heatmap(df.corr())
```

```python
sns.pairplot(df[pairplot_cols], hue="CAND_PTY_AFFILIATION")
```
